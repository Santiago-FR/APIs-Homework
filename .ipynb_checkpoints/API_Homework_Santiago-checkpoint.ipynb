### Analysis

* Maximum Temperatures ten to be much higher closer to the equator (0 degrees latitude) cities below the equator are warmer that those above. this is mostly due to the tilt of the earth this time of the year. There are extremely high temperatures recorded with a large quantity of cities with temperatures ranging from 290 to 310 degrees Fahrenheit. (reference image City_lat_vs_MaxTemp.png). 


* Humidity levels are evenly represented across all latitudes; most cities experience humidity of 60% or higher (reference image City_lat_vs_Humidity.png). 


* There is no strong relationship between latitude and cloudiness. However, a strong band of cities sits at 0, 20, 80, and 100% cloudiness. There are smaller bands of cities at 20 and 40% cloudiness. (reference image City_lat_vs_Cloudines.png). 


* There is no strong relationship between latitude and wind speed. However, in northern hemispheres there are 3 cities exceeding 12.5 mph winds. It is also important to mention that most cities do not exceed 5 mph winds (reference image City_Lat_vs_WindSpeed.png). 


```python
# Dependencies and Setup
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
import requests
import time

# Import API key
from api_keys import api_key

# Incorporated citipy to determine city based on latitude and longitude
from citipy import citipy

# # Output File (CSV) ????
# output_data_file = "output_data/cities.csv" ??????

# Range of latitudes and longitudes
lat_range = (-90, 90)
lng_range = (-180, 180)
```

## Generate Cities List

```python
# List for holding lat_lngs and cities
lat_lngs = []
cities = []

# Create a set of random lat and lng combinations
lats = np.random.uniform(low=-90.000, high=90.000, size=1500)
lngs = np.random.uniform(low=-180.000, high=180.000, size=1500)
lat_lngs = zip(lats, lngs)

# Identify nearest city for each lat, lng combination
for lat_lng in lat_lngs:
    city = citipy.nearest_city(lat_lng[0], lat_lng[1]).city_name
    
    # If the city is unique, then add it to a our cities list
    if city not in cities:
        cities.append(city)

# Print the city count to confirm sufficient count
len(cities)
# next to lines of code were used to reduce jason call size so i woulndt get locked out during debuging.
# new_cities = cities[0:5]
# new_cities
```

### Perform API Calls
* Perform a weather check on each city using a series of successive API calls.
* Include a print log of each city as it'sbeing processed (with the city number and city name).


```python
# configuration information.
url = "http://api.openweathermap.org/data/2.5/weather?"

# partial query URL
query_url = f"{url}appid={api_key}&q="

# print(query_url)
```

```python
# Set lists
name = []
temp_max = []
humidity = []
cloudiness = []
wind_speed = []
lon = []
lat = []
country = []
date = []

# St counter variables (when testing uesed new_cities instead of city)
of = len(cities)
x = 1

# Loop to get info out of json ( while testing used new_cities instead of cities)
for city in cities:
        
        response = requests.get(query_url + city,verify=False).json()
        try:
            temp_max.append(response['main']['temp_max'])
            humidity.append(response['main']['humidity'])
            cloudiness.append(response['clouds']['all'])
            wind_speed.append(response['wind']['speed'])
            date.append(response['dt'])
            country.append(response['sys']['country'])
            lat.append(response['coord']['lat'])
            lon.append(response['coord']['lon'])
            name.append(response['name'])
         

        except KeyError:
            
            print("Cant touch this");
            
# Print log
        try:
            print(f"Processing Record {x} of {of} | {response['name']}")
        except KeyError:
            print("Cant touch this either");
        x += 1
 
```

### Convert Raw Data to DataFrame
* Export the city data into a .csv.
* Display the DataFrame

```python
# Create a dictionary to later convert into a data frame
weather_dict = {
    "City": name,
    "Cloudines":cloudiness,
    "Country": country,
    "Date": date,
    "Humidity": humidity,
    "Latitude": lat,
    "Longitude": lon,
    "Max Temp": temp_max,
    "Wind Speed": wind_speed
}
```

```python
# Turn dictionary into data frame
weather_df = pd.DataFrame(weather_dict)
weather_df.head()
```

```python
# Save csv
weather_csv =weather_df.to_csv (r'C:\Users\La Casa\Desktop\APIs-Homework\weather.csv', index = None, header=True) 

```

#### Latitude vs. Temperature Plot

```python
# check a date from df convert to yyyy/mm/dd
from datetime import datetime
date = int("1552026608")
chart_date = datetime.utcfromtimestamp(date).strftime('%Y-%m-%d')
chart_date
```

```python
c1_X = weather_df["Latitude"]
c1_y = weather_df["Max Temp"]
plt.scatter(c1_X, c1_y, c = "steelblue", marker = "o", alpha = 0.8, edgecolors = "black")
plt.title(f"City Latitude vs. Max Temperature ({chart_date})")
plt.xlabel("Latitude")
plt.ylabel("Max Temperature (F)")
plt.grid(color = "grey", linestyle = "-", linewidth = .5)
plt.savefig("City_Lat_vs_MaxTemp.png")
plt.show()
```
